TIP CALCULATOR

```python
print("Welcome to the Tip calculator")
bill = input("Please enter the bill: ")
tip = input("How much tip you want to give? 10, 12 or 15: ")
tip_per =  float(tip)/100
total_tip=tip_per * float(bill)
total_bill = float(bill) + float(total_tip)
print(total_bill)
persons = input("How many persons will pay the bill: ")
each_person_share = total_bill / float(persons)
final_bill = round(each_person_share,2)
print(f"Each person should pay bill: {final_bill}")

```
