Hnagman Game

```python
import random
stages = ['''
  +---+
  |   |
  O   |
 /|\  |
 / \  |
      |
=========
''', '''
  +---+
  |   |
  O   |
 /|\  |
 /    |
      |
=========
''', '''
  +---+
  |   |
  O   |
 /|\  |
      |
      |
=========
''', '''
  +---+
  |   |
  O   |
 /|   |
      |
      |
=========''', '''
  +---+
  |   |
  O   |
  |   |
      |
      |
=========
''', '''
  +---+
  |   |
  O   |
      |
      |
      |
=========
''', '''
  +---+
  |   |
      |
      |
      |
      |
=========
''']

lives = 6
word_list = ["adavark", "karachi", "lahore"]
chosen_word = random.choice(word_list)
word_length = len(chosen_word)
print(chosen_word)
display = []
#for n in chosen_word:
for n in range(word_length):
  display += "_"
print(display)

end_of_game = False
while not end_of_game:
  guess = input("Please guess a number: ").lower()
  if guess in display:
    print(f"You have already entered letter: {guess}")
  for position in range(word_length):
    letter = chosen_word[position]
    if letter == guess:
      display[position] = letter
  print(display)
  if guess not in chosen_word:
    print(f"You choose letter {guess}, which is not in list")
    lives -= 1
    if lives == 0:
      end_of_game = True
      print("You lose")
  if "_" not in display:
    end_of_game = True
    print("You win")
  print(stages[lives])
```
